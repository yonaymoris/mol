```python
import random
import math

from IPython.display import display, HTML
def show_html(src):
    return display(HTML(src), metadata=dict(isolated=True))
```

2. In this code I tried the example with the stars but used words instead. I thought of the real life comparison with humans where there are multiple "lost" and one "found" and here is what I got.

```python
divs = []
for i in range(200):
    x = random.random() * 90
    y = random.random() * 90
    size = random.randrange(8, 18)
    alpha = random.random()
    this_div = f"<div style='position: absolute; left: {x}%; top: {y}%; font-size: {size}pt;'>lost.</div>"
    divs.append(this_div)

x_found = random.random() * 90
y_found = random.random() * 90
divs.append(f"<div style='position: absolute; left: {x_found}%; top: {y_found}%; font-size: 6pt; color: rgba(245,66,66)'>found.</div>")
html_src = html_tmpl.format(title="Lost and Found", content="".join(divs))
```

```python
show_html(html_src)
```

3. Next I wanted to do something with one of my favorite quotes by Onore de Balzak — "To reach the goal you have to go" but in Russian. In Russian "to reach the goal" literally is translate as "to go to the goal" which makes it a great words play. I decided to place the words like they are forming a path.

```python
sentence = "Чтобы дойти доoo цели, надо прежде всего идти."
s = sentence.split()
print(s)
```

```python
switcher = {
    1: 40,
    2: 45,
    3: 40,
    4: 35,
    5: 40,
    0: 45
}

sentence_divs = []
x = 30
for i in range(0, len(s)):
    y = switcher[(i+1) % 6]
    x = x + len(s[i])
    current_div = f"<div style='position: absolute; left: {x}%; top: {y}%; font-size: 10pt;'>{s[i]}</div>"
    sentence_divs.append(current_div)

html_src = html_tmpl.format(title="To Reach Your Goal", content="".join(sentence_divs))
show_html(html_src)
```

```python
switcher = {
    1: 40,
    2: 45,
    3: 40,
    4: 35,
    5: 40,
    0: 45
}

sentence_divs = []
x = 30
for i in range(0, len(s)):
    y = switcher[(i+1) % 6]
    x = x + len(s[i])
    current_div = f"<div style='position: absolute; left: {x}%; top: {y}%; font-size: 10pt;'>{s[i]}</div>"
    sentence_divs.append(current_div)

html_src = html_tmpl.format(title="To Reach Your Goal", content="".join(sentence_divs))
show_html(html_src)
```

```python
import numpy as np
def mkdiv(content, **kwargs):
    if 'position' not in kwargs:
        kwargs['position'] = 'absolute'
    style_str = ' '.join([": ".join((k.replace('_', '-'), v))+";" for k, v in kwargs.items()])
    return f"<div style='{style_str}'>{content}</div>"
```

5. Next, I tried to create a flower out of Shakespeare's works, just for fun.

```python
words = open("shakespeare.txt").read().split()
divs = []
angle_step = 15
for i, word in enumerate(words):
    angle = i * angle_step
    this_div = mkdiv(word,
                     top=f"50%",
                     left=f"50%",
                     height=f"{12+i*3}pt",
                     font_size="12pt",
#                      border="1px black solid", # uncomment to see the boxes
                     transform=f"translate(-50%, -50%) rotate({angle}deg)")
#     print(this_div)
    divs.append(this_div)
html_src = html_tmpl.format(title="frost spiral", content="".join(divs))
show_html(html_src)
```
